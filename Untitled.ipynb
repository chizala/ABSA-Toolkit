```python
import xml.etree.ElementTree as ET, getopt, logging, sys, random, re, copy, os
import numpy as np
from collections import Counter
import operator
from pystruct.models import ChainCRF
from pystruct.learners import FrankWolfeSSVM
from sklearn.linear_model import LogisticRegression
import nltk
import pandas as pd
from xml.sax.saxutils import escape

def prepare_df(fname):
    sentences_array =[]
    categories=[]
    ote=[]
    polarity=[]
    reviews = ET.parse(fname).getroot().findall('Review')
    i = 1
    for review in reviews:
        sentences = review[0] #get the sentences
       
        for sentence in sentences:
            
            if (len(sentence) > 1):

                opinions = sentence[1] #getting the opinions field
                
                if ( len(opinions) > 0): #check if there are aspects 
                   
                    t = sentence[0].text #getting the text
                    cat=''
                    aspect=''
                    pol=''
                    for opinion in opinions:
                        if opinion.attrib['category'].lower() != 'notrelated':
                                if opinion.attrib['category'].lower() == 'other':
                                    cat = cat + ' '+ 'hotel'
                                else:
                                    cat = cat + ' '+ opinion.attrib['category'].lower()
                                aspect = aspect + ';'+ opinion.attrib['target'].lower()
                                pol = pol+' '+opinion.attrib['polarity'].lower()

                    if cat != '':
                        ote.append(aspect) #store the category
                        categories.append(cat)
                        polarity.append(pol)
                        sentences_array.append(sentence[0].text)

    return sentences_array,categories,ote,polarity
```

```python
sentences_array,categories,ote,polarity = prepare_df('data/hotels/hotel_train.xml')
train_SF = pd.DataFrame({'text' : sentences_array,'category' : categories,'aspect term':ote,'polarity':polarity})
```

```python
sentences_array,categories,ote,polarity = prepare_df('data/hotels/hotel_test-1.xml')
test_SF = pd.DataFrame({'text' : sentences_array,'category' : categories,'aspect term':ote,'polarity':polarity})
```
